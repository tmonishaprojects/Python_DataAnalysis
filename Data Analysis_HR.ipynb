# Data Analysis & Data Visualization

1) Count of employees in each department

```python
Count_Emp=data.groupby('Department')['Department'].count()
```

```python
Count_Emp
```

```python
explode=[0,0,0.1,0,0,0]
palette_color=sns.color_palette('bright')
label=['Admin Offices','Executive Office','IT/IS','Production','Sales','Software Engineering']
plt.pie(Count_Emp,autopct='%.0f%%',explode=explode,pctdistance=0.85,labels=label)
```

2) Average salary of employess in each depatment

```python
Avg_Salary=data.groupby('Department')['Salary'].mean()
```

```python
Avg_Salary.round(2)
```

```python
explode=[0,0,0.1,0,0,0]
palette_color=sns.color_palette('bright')
label=['Admin Offices','Executive Office','IT/IS','Production','Sales','Software Engineering']
plt.pie(Avg_Salary,autopct='%.0f%%',explode=explode,pctdistance=0.85,labels=label)
```

3) Calculate the average of salary of employees based on gender.

```python
sex_salary=data.groupby(['Department','Sex'],as_index=False).Salary.mean()
```

```python
sex_salary
```

```python
sns.barplot(x='Department',y='Salary',hue='Sex',data=sex_salary)
```

4) Most popular RecruitmentSource of the company based on department.

```python
Recruitmentsource=data.groupby(['Department','RecruitmentSource']).size().reset_index(name='counts')
```

```python
Recruitmentsource
```

```python
sns.barplot(x='RecruitmentSource',y='counts',hue='Department',data=Recruitmentsource)
```

5) Total Number of working days for terminated employee.

```python
T_year=int(input("Enter the Terminated year: "))
T_month=int(input("Enter the Terminated month: "))
T_day=int(input("Enter the Terminated day: "))
T_Date=date(T_year,T_month,T_day)
```

```python
business_days=pd.bdate_range(Start_Date,T_Date)
```
